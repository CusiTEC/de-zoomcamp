````python
df = pd.read_csv('green_tripdata_2019-09.csv.gz', nrows=449065)
````

````python
df.head()
````

````python
df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
````

````python
df.to_sql(name='green_tripdata_trip', con=engine, index=False)
````

````python
query = """
SELECT * FROM green_tripdata_trip LIMIT 10
"""

pd.read_sql(query, con=engine)
````

````python
query = """
SELECT COUNT(*) FROM green_tripdata_trip
"""

pd.read_sql(query, con=engine)
````

````python
query = """
SELECT COUNT(*) AS total_trips
FROM green_tripdata_trip
WHERE lpep_pickup_datetime >= '2019-09-18' AND lpep_pickup_datetime < '2019-09-19'
AND lpep_dropoff_datetime >= '2019-09-18' AND lpep_dropoff_datetime < '2019-09-19';
"""

pd.read_sql(query, con=engine)
````

````python
query = """
SELECT
  date(lpep_pickup_datetime),
  MAX(trip_distance) as max_trip_distance
FROM green_tripdata_trip
GROUP BY date(lpep_pickup_datetime)
ORDER BY max_trip_distance DESC;
"""

pd.read_sql(query, con=engine)
````

````python
query = """
SELECT
  z."Borough",
  SUM(t."total_amount")
FROM green_tripdata_trip t 
INNER JOIN zones z
ON t."PULocationID" = z."LocationID"
WHERE date(lpep_pickup_datetime) = '2019-09-18'
AND z."Borough" <> 'Unknown'
GROUP BY z."Borough"
HAVING SUM(t."total_amount") >=50000;
"""

pd.read_sql(query, con=engine)
````

````python
query = """
SELECT
  z2."Zone",
  MAX(t."tip_amount") AS max_tip
FROM green_tripdata_trip t
INNER JOIN zones z
ON t."PULocationID" = z."LocationID"
INNER JOIN zones z2
ON t."DOLocationID" = z2."LocationID"
WHERE date(t."lpep_pickup_datetime") BETWEEN '2019-09-01' AND '2019-09-30'
AND z."Zone" = 'Astoria'
GROUP BY z2."Zone"
ORDER BY max_tip DESC
LIMIT 1;
"""

pd.read_sql(query, con=engine)
````
